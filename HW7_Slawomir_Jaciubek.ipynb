Task 6

Three dice are thrown. What is more likely: to get a total of 11 or 12 points?

Solve programmatically only

```python
import itertools
val = [1, 2, 3, 4, 5, 6]
permutation_set = itertools.permutations(val,3)
n=0
n_11=0
n_12=0

for elem in permutation_set:
    print(elem)
    n+=1
    if sum(elem) == 11:
        n_11+=1
    if sum(elem) == 12:
        n_12+=1

print(n)
print(f"11 point occurance: {n_11} times")
print(f"12 point occurance: {n_12} times")
print("occurance of 12 and 11 points has the same probability")
```
